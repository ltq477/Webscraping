```python
import bs4 as BeautifulSoup
import requests
import urllib
from fake_useragent import UserAgent
from bs4 import BeautifulSoup
```

```python
ua = UserAgent()
header = {'user-agent':ua.chrome}
```

```python
url = requests.get('https://en.wikipedia.org/wiki/List_of_Fairy_Tail_episodes',headers=header).text
soup = BeautifulSoup(url, 'lxml')
print(soup.prettify)
```

```python
list = soup('td')

for x in list:
    if len(list) >= 1:
        print(x.get_text())
        
```
