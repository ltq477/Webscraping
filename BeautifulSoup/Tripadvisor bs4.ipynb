```python
import bs4 as BeautifulSoup
import requests
import pandas as pd
import urllib
from fake_useragent import UserAgent
from bs4 import BeautifulSoup
```

```python
ua = UserAgent()
header = {'user-agent':ua.chrome}
```

```python
url = requests.get('https://www.tripadvisor.com/HotelsNear-g298184-qHND-Tokyo_Tokyo_Prefecture_Kanto.html', headers=header).text
```

```python
soup = BeautifulSoup(url,'lxml')
print(soup.prettify())
```

# Scraping

```python
Hotel_name = soup('a', {'target':'_blank'})

for names in Hotel_name:
    print(names.get_text())
```

```python
Distance = soup('b')

for dist in Distance:
    if len(dist.contents) <= 15:
        print(dist.contents[0])
```
